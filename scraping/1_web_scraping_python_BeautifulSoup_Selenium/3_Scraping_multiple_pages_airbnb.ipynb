```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.airbnb.co.id/s/Jepang/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&price_filter_input_type=0&price_filter_num_nights=3&date_picker_type=calendar&checkin=2023-01-20&checkout=2023-01-23&source=structured_search_input_header&search_type=filter_change&place_id=ChIJLxl_1w9OZzQRRFJmfNR1QvU&query=Jepang&federated_search_session_id=4795016e-327d-4766-be45-6d7ed151ab38&pagination_search=true&cursor=eyJzZWN0aW9uX29mZnNldCI6MCwiaXRlbXNfb2Zmc2V0IjowLCJ2ZXJzaW9uIjoxfQ%3D%3D'
```

```python
requests.get(url)
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
# soup
```

```python
## next page
# kenapa pakai dictionary class bukan class_ (?)) untuk mengambil atribut yang ilegal adalah kita harus memasukkannya ke dalam kamus ini karena tidak umum digunakan.
# Jika kita menggunakan class, ingat, kita bisa meletakkan class underscore. tetapi karena ada label itu, kita harus meletakkan dalam bentuk kamus.
next_page = soup.find('a', {'aria-label': 'Berikutnya'}).get('href')
next_page_full = 'https://www.airbnb.co.id/'+next_page
next_page_full

url_each_page = next_page_full
request_each_page = requests.get(url_each_page)
soup_ep = BeautifulSoup(request_each_page.text, 'lxml')
soup_ep
```
